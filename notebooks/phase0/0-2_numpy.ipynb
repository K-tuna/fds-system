## 4. ì—°ì‚°

```python
# ğŸ“š ì§‘ê³„ ì—°ì‚°
arr = np.array([10, 20, 30, 40, 50])

print(f"sum:  {arr.sum()}")   # í•©ê³„
print(f"mean: {arr.mean()}") # í‰ê· 
print(f"std:  {arr.std()}")  # í‘œì¤€í¸ì°¨
print(f"max:  {arr.max()}")  # ìµœëŒ€
print(f"min:  {arr.min()}")  # ìµœì†Œ
```

```python
# ğŸ“š axis ê°œë… (2ì°¨ì›)
arr2d = np.array([[1, 2, 3],
                  [4, 5, 6]])

print("ì „ì²´:")
print(arr2d)
print()
print(f"ì „ì²´ í•©: {arr2d.sum()}")           # 21
print(f"í–‰ë³„ í•© (axis=1): {arr2d.sum(axis=1)}")  # [6, 15]
print(f"ì—´ë³„ í•© (axis=0): {arr2d.sum(axis=0)}")  # [5, 7, 9]
```

ê±°ë˜ ê¸ˆì•¡ ë°ì´í„°ì˜ í†µê³„ë¥¼ ê³„ì‚°í•˜ì„¸ìš”.

```python
# ğŸ’» ì‹¤ìŠµ 3
amounts = np.array([15000, 3200, 89000, 4500, 120000, 7800, 55000])

# TODO: ì´ ê±°ë˜ì•¡
total = amounts.sum()

# TODO: í‰ê·  ê±°ë˜ì•¡
avg = amounts.mean()

# TODO: ìµœëŒ€ ê±°ë˜ì•¡
max_amt = amounts.max()

print(f"ì´ ê±°ë˜ì•¡: {total:,}ì›")
print(f"í‰ê·  ê±°ë˜ì•¡: {avg:,.0f}ì›")
print(f"ìµœëŒ€ ê±°ë˜ì•¡: {max_amt:,}ì›")
```

## 5. ì¡°ê±´ í•„í„°ë§

```python
# ğŸ“š ë¶ˆë¦¬ì–¸ ë°°ì—´
arr = np.array([10, 25, 30, 5, 40])

# ì¡°ê±´ ì ìš© â†’ ë¶ˆë¦¬ì–¸ ë°°ì—´
mask = arr > 20
print("arr > 20:", mask)  # [False, True, True, False, True]

# ë¶ˆë¦¬ì–¸ìœ¼ë¡œ í•„í„°ë§
filtered = arr[mask]
print("í•„í„°ë§:", filtered)  # [25, 30, 40]
```

```python
# ğŸ“š np.where
arr = np.array([10, 25, 30, 5, 40])

# ì¡°ê±´ì— ë”°ë¼ ê°’ ë³€ê²½
result = np.where(arr > 20, "high", "low")
print(result)  # ['low', 'high', 'high', 'low', 'high']
```

### ğŸ’» ì‹¤ìŠµ 4: ì¡°ê±´ í•„í„°ë§ (ì‚¬ê¸° ê±°ë˜ ì°¾ê¸°)

100,000ì› ì´ìƒ ê±°ë˜ë¥¼ ê³ ì•¡ ê±°ë˜ë¡œ ë¶„ë¥˜í•˜ì„¸ìš”.

```python
# ğŸ’» ì‹¤ìŠµ 4
transactions = np.array([15000, 250000, 8900, 120000, 45000, 980000, 3200])

# TODO: 100,000ì› ì´ìƒì¸ ê±°ë˜ë§Œ í•„í„°ë§
high_value = transactions[transactions >= 100000]

# TODO: ê³ ì•¡ ê±°ë˜ ê°œìˆ˜
high_count = len(high_value)

# TODO: np.whereë¡œ "ê³ ì•¡"/"ì¼ë°˜" ë¼ë²¨ë§
labels = np.where(transactions >= 100000, "ê³ ì•¡", "ì¼ë°˜")

print("ê³ ì•¡ ê±°ë˜:", high_value)
print("ê³ ì•¡ ê±°ë˜ ìˆ˜:", high_count)
print("ë¼ë²¨:", labels)
```

```python
# ì²´í¬í¬ì¸íŠ¸ 4
assert list(high_value) == [250000, 120000, 980000], "ê³ ì•¡ ê±°ë˜"
assert high_count == 3, "ê³ ì•¡ ê±°ë˜ ìˆ˜"
assert list(labels) == ["ì¼ë°˜", "ê³ ì•¡", "ì¼ë°˜", "ê³ ì•¡", "ì¼ë°˜", "ê³ ì•¡", "ì¼ë°˜"], "ë¼ë²¨"

print("âœ… ì²´í¬í¬ì¸íŠ¸ 4 í†µê³¼!")
```

## 6. ë¸Œë¡œë“œìºìŠ¤íŒ…

```python
# ğŸ“š ë¸Œë¡œë“œìºìŠ¤íŒ…
# ìŠ¤ì¹¼ë¼ ì—°ì‚°ì´ ì „ì²´ ë°°ì—´ì— ì ìš©ë¨

arr = np.array([10, 20, 30, 40, 50])

print("ì›ë³¸:", arr)
print("* 2:", arr * 2)
print("+ 5:", arr + 5)
print("/ 10:", arr / 10)
```

```python
# ğŸ“š ì •ê·œí™” ì˜ˆì‹œ (Min-Max Scaling)
# (x - min) / (max - min) â†’ 0~1 ë²”ìœ„ë¡œ ë³€í™˜

data = np.array([100, 200, 300, 400, 500])

min_val = data.min()
max_val = data.max()
normalized = (data - min_val) / (max_val - min_val)

print("ì›ë³¸:", data)
print("ì •ê·œí™”:", normalized)
```

### ğŸ’» ì‹¤ìŠµ 5: ì •ê·œí™”

ê±°ë˜ ê¸ˆì•¡ì„ 0~1 ë²”ìœ„ë¡œ ì •ê·œí™”í•˜ì„¸ìš”.

```python
# ğŸ’» ì‹¤ìŠµ 5
amounts = np.array([10000, 50000, 30000, 80000, 20000])

# TODO: Min-Max ì •ê·œí™” (0~1)
normalized = (amounts - amounts.min()) / (amounts.max() - amounts.min())

print("ì›ë³¸:", amounts)
print("ì •ê·œí™”:", normalized)
```

```python
# ì²´í¬í¬ì¸íŠ¸ 5
expected = np.array([0.0, 0.571, 0.286, 1.0, 0.143])
assert np.allclose(normalized, expected, atol=0.01), "ì •ê·œí™” ê²°ê³¼"

print("âœ… ì²´í¬í¬ì¸íŠ¸ 5 í†µê³¼!")
```
