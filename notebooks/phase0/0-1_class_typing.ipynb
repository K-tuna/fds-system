```python
class BankAccount:
    def __init__(self, owner, balance=0):
        print(f"ê³„ì¢Œ ìƒì„±: {owner}")
        self.owner = owner
        self.balance = balance
    
    def deposit(self, amount):
        self.balance += amount
        return self.balance
    
    def withdraw(self, amount):
        if amount > self.balance:
            print("ì”ì•¡ ë¶€ì¡±!")
            return self.balance
        self.balance -= amount
        return self.balance

# ê°ì²´ ìƒì„± (ì´ë•Œ __init__ í˜¸ì¶œ)
account = BankAccount("í™ê¸¸ë™", 10000)
print(f"ì”ì•¡: {account.balance}")

account.deposit(5000)
print(f"ì…ê¸ˆ í›„: {account.balance}")

account.withdraw(3000)
print(f"ì¶œê¸ˆ í›„: {account.balance}")
```

## 3. sklearn íŒ¨í„´

### ğŸ“š sklearnì´ ì™œ í´ë˜ìŠ¤ì¸ê°€?

ëª¨ë¸ì€ **ìƒíƒœ(í•™ìŠµëœ íŒŒë¼ë¯¸í„°)**ë¥¼ ì €ì¥í•´ì•¼ í•¨

- `fit()`: í•™ìŠµ â†’ íŒŒë¼ë¯¸í„° ì €ì¥
- `predict()`: ì €ì¥ëœ íŒŒë¼ë¯¸í„°ë¡œ ì˜ˆì¸¡

```python
# ğŸ“š sklearnì´ ì™œ í´ë˜ìŠ¤ì¸ê°€?
# ëª¨ë¸ì€ í•™ìŠµëœ íŒŒë¼ë¯¸í„°ë¥¼ ì €ì¥í•´ì•¼ í•¨

class SimpleModel:
    def __init__(self):
        self.is_fitted = False
        self.mean_value = None
    
    def fit(self, X, y):
        """í•™ìŠµ: í‰ê· ê°’ ì €ì¥"""
        self.mean_value = sum(y) / len(y)
        self.is_fitted = True
        return self
    
    def predict(self, X):
        """ì˜ˆì¸¡: í•­ìƒ í‰ê· ê°’ ë°˜í™˜"""
        if not self.is_fitted:
            raise ValueError("ë¨¼ì € fit()ì„ í˜¸ì¶œí•˜ì„¸ìš”")
        return [self.mean_value] * len(X)

model = SimpleModel()
model.fit([1,2,3], [10, 20, 30])
print(model.predict([4, 5]))  # [20.0, 20.0]
```
