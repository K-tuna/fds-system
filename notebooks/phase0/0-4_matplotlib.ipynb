## 5. ê·¸ë˜í”„ ê¾¸ë¯¸ê¸°

```python
# ğŸ“š ê·¸ë˜í”„ ê¾¸ë¯¸ê¸° ì¢…í•©
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.figure(figsize=(10, 6))  # í¬ê¸° ì„¤ì •
plt.plot(x, y, 'b-', linewidth=2)

plt.title('Sine Wave', fontsize=16)  # ì œëª©
plt.xlabel('X axis', fontsize=12)     # xì¶• ë¼ë²¨
plt.ylabel('Y axis', fontsize=12)     # yì¶• ë¼ë²¨

plt.xlim(0, 10)   # xì¶• ë²”ìœ„
plt.ylim(-1.5, 1.5)  # yì¶• ë²”ìœ„

plt.grid(True, alpha=0.3)  # ê·¸ë¦¬ë“œ
plt.show()
```

## 6. Subplot

```python
# ğŸ“š subplot: ì—¬ëŸ¬ ê·¸ë˜í”„ í•œë²ˆì—
fig, axes = plt.subplots(2, 2, figsize=(10, 8))  # 2x2 ê·¸ë¦¬ë“œ

# ì¢Œìƒ
axes[0, 0].plot([1, 2, 3], [1, 2, 3])
axes[0, 0].set_title('Line')

# ìš°ìƒ
axes[0, 1].bar(['A', 'B', 'C'], [3, 5, 2])
axes[0, 1].set_title('Bar')

# ì¢Œí•˜
axes[1, 0].hist(np.random.randn(100), bins=20)
axes[1, 0].set_title('Histogram')

# ìš°í•˜
axes[1, 1].scatter([1, 2, 3, 4], [1, 4, 2, 3])
axes[1, 1].set_title('Scatter')

plt.tight_layout()  # ê²¹ì¹¨ ë°©ì§€
plt.show()
```
