```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# ÌïúÍ∏Ä Ìè∞Ìä∏ ÏÑ§Ï†ï (Windows: ÎßëÏùÄ Í≥†Îîï)
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False  # ÎßàÏù¥ÎÑàÏä§ Íπ®Ïßê Î∞©ÏßÄ

# ÎÖ∏Ìä∏Î∂ÅÏóêÏÑú Í∑∏ÎûòÌîÑ Î∞îÎ°ú ÌëúÏãú
%matplotlib inline

print("‚úÖ ÌïúÍ∏Ä Ìè∞Ìä∏ ÏÑ§Ï†ï ÏôÑÎ£å")
```

## 1. Í∏∞Ï¥à

```python
# üìö Í∞ÄÏû• Í∞ÑÎã®Ìïú Í∑∏ÎûòÌîÑ
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.plot(x, y)
plt.show()
```

```python
# üìö figureÏôÄ axes
fig, ax = plt.subplots()  # figure: ÎèÑÌôîÏßÄ, ax: Í∑∏ÎûòÌîÑ ÏòÅÏó≠
ax.plot(x, y)
plt.show()
```

## 2. ÏÑ† Í∑∏ÎûòÌîÑ

```python
# üìö ÏÑ† Í∑∏ÎûòÌîÑ ÏòµÏÖò
x = np.arange(0, 10, 0.1)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, 'b-', label='sin')   # ÌååÎûÄ Ïã§ÏÑ†
plt.plot(x, y2, 'r--', label='cos')  # Îπ®Í∞Ñ Ï†êÏÑ†
plt.legend()  # Î≤îÎ°Ä ÌëúÏãú
plt.show()
```

```python
# üìö ÏÉâÏÉÅ, Ïä§ÌÉÄÏùº ÏòµÏÖò
# ÏÉâÏÉÅ: b(ÌååÎûë), r(Îπ®Í∞ï), g(Ï¥àÎ°ù), k(Í≤ÄÏ†ï), c(Ï≤≠Î°ù), m(ÎßàÏ††ÌÉÄ), y(ÎÖ∏Îûë)
# Ïä§ÌÉÄÏùº: -(Ïã§ÏÑ†), --(Ï†êÏÑ†), :(Ï†ê), -.(Ï†êÏÑ†+Ï†ê)
# ÎßàÏª§: o(Ïõê), s(ÏÇ¨Í∞Å), ^(ÏÇºÍ∞Å), *(Î≥Ñ)

x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

plt.plot(x, y, 'go--', linewidth=2, markersize=10)  # Ï¥àÎ°ù Ïõê Ï†êÏÑ†
plt.show()
```

```python
# üìö ÌûàÏä§ÌÜ†Í∑∏Îû® Í∏∞Ï¥à
data = np.random.randn(1000)  # Ï†ïÍ∑úÎ∂ÑÌè¨ ÎÇúÏàò

plt.hist(data, bins=30)  # bins: ÎßâÎåÄ Í∞úÏàò
plt.show()
```

## 5. Í∑∏ÎûòÌîÑ Íæ∏ÎØ∏Í∏∞

```python
# üìö Í∑∏ÎûòÌîÑ Íæ∏ÎØ∏Í∏∞ Ï¢ÖÌï©
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.figure(figsize=(10, 6))  # ÌÅ¨Í∏∞ ÏÑ§Ï†ï
plt.plot(x, y, 'b-', linewidth=2)

plt.title('Sine Wave', fontsize=16)  # Ï†úÎ™©
plt.xlabel('X axis', fontsize=12)     # xÏ∂ï ÎùºÎ≤®
plt.ylabel('Y axis', fontsize=12)     # yÏ∂ï ÎùºÎ≤®

plt.xlim(0, 10)   # xÏ∂ï Î≤îÏúÑ
plt.ylim(-1.5, 1.5)  # yÏ∂ï Î≤îÏúÑ

plt.grid(True, alpha=0.3)  # Í∑∏Î¶¨Îìú
plt.show()
```

## 6. Subplot

```python
# üìö subplot: Ïó¨Îü¨ Í∑∏ÎûòÌîÑ ÌïúÎ≤àÏóê
fig, axes = plt.subplots(2, 2, figsize=(10, 8))  # 2x2 Í∑∏Î¶¨Îìú

# Ï¢åÏÉÅ
axes[0, 0].plot([1, 2, 3], [1, 2, 3])
axes[0, 0].set_title('Line')

# Ïö∞ÏÉÅ
axes[0, 1].bar(['A', 'B', 'C'], [3, 5, 2])
axes[0, 1].set_title('Bar')

# Ï¢åÌïò
axes[1, 0].hist(np.random.randn(100), bins=20)
axes[1, 0].set_title('Histogram')

# Ïö∞Ìïò
axes[1, 1].scatter([1, 2, 3, 4], [1, 4, 2, 3])
axes[1, 1].set_title('Scatter')

plt.tight_layout()  # Í≤πÏπ® Î∞©ÏßÄ
plt.show()
```
