## 4. groupby

```python
# ğŸ“š groupby ê¸°ì´ˆ
df = pd.DataFrame({
    'category': ['A', 'B', 'A', 'B', 'A'],
    'value': [10, 20, 30, 40, 50]
})

# ì¹´í…Œê³ ë¦¬ë³„ í•©ê³„
print("ì¹´í…Œê³ ë¦¬ë³„ í•©ê³„:")
print(df.groupby('category')['value'].sum())
print()

# ì¹´í…Œê³ ë¦¬ë³„ í‰ê· 
print("ì¹´í…Œê³ ë¦¬ë³„ í‰ê· :")
print(df.groupby('category')['value'].mean())
```

```python
# ğŸ“š agg()ë¡œ ì—¬ëŸ¬ ì§‘ê³„
result = df.groupby('category')['value'].agg(['sum', 'mean', 'count'])
result
```

### ğŸ’» ì‹¤ìŠµ 4: groupby

ê±°ë˜ ë°ì´í„°ë¥¼ ì‚¬ê¸° ì—¬ë¶€ë³„ë¡œ ì§‘ê³„í•˜ì„¸ìš”.

```python
# ğŸ’» ì‹¤ìŠµ 4
# TODO: is_fraudë³„ ê±°ë˜ ê±´ìˆ˜
fraud_count = transactions.groupby('is_fraud')['transaction_id'].count()

# TODO: is_fraudë³„ í‰ê·  ê±°ë˜ì•¡
fraud_avg = transactions.groupby('is_fraud')['amount'].mean()

print("ì‚¬ê¸° ì—¬ë¶€ë³„ ê±´ìˆ˜:")
print(fraud_count)
print()
print("ì‚¬ê¸° ì—¬ë¶€ë³„ í‰ê·  ê±°ë˜ì•¡:")
print(fraud_avg)
```

```python
# ì²´í¬í¬ì¸íŠ¸ 4
assert fraud_count[False] == 4, "ì •ìƒ ê±°ë˜ 4ê±´"
assert fraud_count[True] == 1, "ì‚¬ê¸° ê±°ë˜ 1ê±´"
assert fraud_avg[True] == 120000, "ì‚¬ê¸° ê±°ë˜ í‰ê· "

print("âœ… ì²´í¬í¬ì¸íŠ¸ 4 í†µê³¼!")
```

## 5. merge (ì¡°ì¸)

```python
# ğŸ“š merge ê¸°ì´ˆ
# ê±°ë˜ í…Œì´ë¸”
tx = pd.DataFrame({
    'tx_id': [1, 2, 3],
    'user_id': [101, 102, 101],
    'amount': [5000, 8000, 3000]
})

# ì‚¬ìš©ì í…Œì´ë¸”
users = pd.DataFrame({
    'user_id': [101, 102, 103],
    'name': ['í™ê¸¸ë™', 'ê¹€ì² ìˆ˜', 'ì´ì˜í¬']
})

print("ê±°ë˜:")
print(tx)
print()
print("ì‚¬ìš©ì:")
print(users)
```

```python
# ğŸ“š INNER JOIN (êµì§‘í•©)
merged = pd.merge(tx, users, on='user_id', how='inner')
print("INNER JOIN:")
merged
```

```python
# ğŸ“š LEFT JOIN (ì™¼ìª½ ê¸°ì¤€)
merged = pd.merge(tx, users, on='user_id', how='left')
print("LEFT JOIN:")
merged
```

### ğŸ’» ì‹¤ìŠµ 5: merge

ê±°ë˜ì™€ ì¹´ë“œ ì •ë³´ë¥¼ ì¡°ì¸í•˜ì„¸ìš”.

```python
# ğŸ’» ì‹¤ìŠµ 5
# ê±°ë˜ ë°ì´í„°
tx = pd.DataFrame({
    'tx_id': [1, 2, 3, 4],
    'card_id': ['C001', 'C002', 'C001', 'C003'],
    'amount': [10000, 25000, 5000, 80000]
})

# ì¹´ë“œ ë°ì´í„°
cards = pd.DataFrame({
    'card_id': ['C001', 'C002'],
    'card_type': ['ì‹ ìš©', 'ì²´í¬']
})

# TODO: LEFT JOINìœ¼ë¡œ ê±°ë˜ì— ì¹´ë“œ íƒ€ì… ë¶™ì´ê¸°
result = pd.merge(tx, cards, on='card_id', how='left')

result
```

```python
# ì²´í¬í¬ì¸íŠ¸ 5
assert len(result) == 4, "4ê±´ ìœ ì§€"
assert 'card_type' in result.columns, "card_type ì»¬ëŸ¼ ì¡´ì¬"
assert pd.isna(result.loc[result['card_id'] == 'C003', 'card_type'].values[0]), "C003ì€ NaN"

print("âœ… ì²´í¬í¬ì¸íŠ¸ 5 í†µê³¼!")
```

## 6. ê²°ì¸¡ì¹˜ ì²˜ë¦¬

```python
# ğŸ“š ê²°ì¸¡ì¹˜ í™•ì¸
df = pd.DataFrame({
    'A': [1, 2, np.nan, 4],
    'B': [np.nan, 2, 3, 4],
    'C': [1, 2, 3, 4]
})

print("ë°ì´í„°:")
print(df)
print()
print("ê²°ì¸¡ì¹˜ ê°œìˆ˜:")
print(df.isna().sum())
```

```python
# ğŸ“š fillna(): ê²°ì¸¡ì¹˜ ì±„ìš°ê¸°
filled = df.fillna(0)
print("0ìœ¼ë¡œ ì±„ìš°ê¸°:")
print(filled)
print()

# í‰ê· ìœ¼ë¡œ ì±„ìš°ê¸°
filled_mean = df.fillna(df.mean())
print("í‰ê· ìœ¼ë¡œ ì±„ìš°ê¸°:")
print(filled_mean)
```

```python
# ğŸ“š dropna(): ê²°ì¸¡ì¹˜ í–‰ ì œê±°
dropped = df.dropna()
print("ê²°ì¸¡ì¹˜ í–‰ ì œê±°:")
print(dropped)
```

### ğŸ’» ì‹¤ìŠµ 6: ê²°ì¸¡ì¹˜ ì²˜ë¦¬

ê±°ë˜ ë°ì´í„°ì˜ ê²°ì¸¡ì¹˜ë¥¼ ì²˜ë¦¬í•˜ì„¸ìš”.

```python
# ğŸ’» ì‹¤ìŠµ 6
df = pd.DataFrame({
    'tx_id': [1, 2, 3, 4, 5],
    'amount': [10000, np.nan, 5000, np.nan, 8000],
    'category': ['ì‹í’ˆ', 'ì˜ë¥˜', None, 'ì‹í’ˆ', 'ì˜ë¥˜']
})

print("ì›ë³¸:")
print(df)
print()

# TODO: amount ê²°ì¸¡ì¹˜ ê°œìˆ˜
amount_na_count = df['amount'].isna().sum()

# TODO: amount ê²°ì¸¡ì¹˜ë¥¼ í‰ê· ìœ¼ë¡œ ì±„ìš°ê¸°
df_filled = df.copy()
df_filled['amount'] = df['amount'].fillna(df['amount'].mean())  # TODO: fillna ì‚¬ìš©

# TODO: category ê²°ì¸¡ì¹˜ë¥¼ 'ê¸°íƒ€'ë¡œ ì±„ìš°ê¸°
df_filled['category'] = df['category'].fillna('ê¸°íƒ€')  # TODO: fillna ì‚¬ìš©

print(f"amount ê²°ì¸¡ì¹˜: {amount_na_count}ê°œ")
print()
print("ì±„ìš´ í›„:")
print(df_filled)
```

```python
# ì²´í¬í¬ì¸íŠ¸ 6
assert amount_na_count == 2, "amount ê²°ì¸¡ì¹˜ 2ê°œ"
assert df_filled['amount'].isna().sum() == 0, "amount ê²°ì¸¡ì¹˜ ì—†ì–´ì•¼ í•¨"
assert df_filled['category'].isna().sum() == 0, "category ê²°ì¸¡ì¹˜ ì—†ì–´ì•¼ í•¨"
assert df_filled.loc[2, 'category'] == 'ê¸°íƒ€', "category ê²°ì¸¡ì¹˜ëŠ” 'ê¸°íƒ€'"

print("âœ… ì²´í¬í¬ì¸íŠ¸ 6 í†µê³¼!")
```
