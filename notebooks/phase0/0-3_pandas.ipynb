## 4. groupby

```python
# üìö groupby Í∏∞Ï¥à
df = pd.DataFrame({
    'category': ['A', 'B', 'A', 'B', 'A'],
    'value': [10, 20, 30, 40, 50]
})

# Ïπ¥ÌÖåÍ≥†Î¶¨Î≥Ñ Ìï©Í≥Ñ
print("Ïπ¥ÌÖåÍ≥†Î¶¨Î≥Ñ Ìï©Í≥Ñ:")
print(df.groupby('category')['value'].sum())
print()

# Ïπ¥ÌÖåÍ≥†Î¶¨Î≥Ñ ÌèâÍ∑†
print("Ïπ¥ÌÖåÍ≥†Î¶¨Î≥Ñ ÌèâÍ∑†:")
print(df.groupby('category')['value'].mean())
```

```python
# üìö agg()Î°ú Ïó¨Îü¨ ÏßëÍ≥Ñ
result = df.groupby('category')['value'].agg(['sum', 'mean', 'count'])
result
```

## 5. merge (Ï°∞Ïù∏)

```python
# üìö merge Í∏∞Ï¥à
# Í±∞Îûò ÌÖåÏù¥Î∏î
tx = pd.DataFrame({
    'tx_id': [1, 2, 3],
    'user_id': [101, 102, 101],
    'amount': [5000, 8000, 3000]
})

# ÏÇ¨Ïö©Ïûê ÌÖåÏù¥Î∏î
users = pd.DataFrame({
    'user_id': [101, 102, 103],
    'name': ['ÌôçÍ∏∏Îèô', 'ÍπÄÏ≤†Ïàò', 'Ïù¥ÏòÅÌù¨']
})

print("Í±∞Îûò:")
print(tx)
print()
print("ÏÇ¨Ïö©Ïûê:")
print(users)
```

```python
# üìö INNER JOIN (ÍµêÏßëÌï©)
merged = pd.merge(tx, users, on='user_id', how='inner')
print("INNER JOIN:")
merged
```

```python
# üìö LEFT JOIN (ÏôºÏ™Ω Í∏∞Ï§Ä)
merged = pd.merge(tx, users, on='user_id', how='left')
print("LEFT JOIN:")
merged
```

## 6. Í≤∞Ï∏°Ïπò Ï≤òÎ¶¨

```python
# üìö Í≤∞Ï∏°Ïπò ÌôïÏù∏
df = pd.DataFrame({
    'A': [1, 2, np.nan, 4],
    'B': [np.nan, 2, 3, 4],
    'C': [1, 2, 3, 4]
})

print("Îç∞Ïù¥ÌÑ∞:")
print(df)
print()
print("Í≤∞Ï∏°Ïπò Í∞úÏàò:")
print(df.isna().sum())
```

```python
# üìö fillna(): Í≤∞Ï∏°Ïπò Ï±ÑÏö∞Í∏∞
filled = df.fillna(0)
print("0ÏúºÎ°ú Ï±ÑÏö∞Í∏∞:")
print(filled)
print()

# ÌèâÍ∑†ÏúºÎ°ú Ï±ÑÏö∞Í∏∞
filled_mean = df.fillna(df.mean())
print("ÌèâÍ∑†ÏúºÎ°ú Ï±ÑÏö∞Í∏∞:")
print(filled_mean)
```

```python
# üìö dropna(): Í≤∞Ï∏°Ïπò Ìñâ Ï†úÍ±∞
dropped = df.dropna()
print("Í≤∞Ï∏°Ïπò Ìñâ Ï†úÍ±∞:")
print(dropped)
```

### üíª Ïã§Ïäµ 6: Í≤∞Ï∏°Ïπò Ï≤òÎ¶¨
